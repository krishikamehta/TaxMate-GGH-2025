```python
# Function to apply tax regulations
def apply_tax_regulations(financial_df, regulations_df):
    """Calculates estimated tax based on income and deductions."""

    def calculate_tax(user_income, deductions, standard_deduction):
        """Applies tax slab rules to determine tax amount."""
        if user_income <= 300000:
            tax_rate = 0.0
        elif user_income <= 700000:
            tax_rate = 0.05
        elif user_income <= 1000000:
            tax_rate = 0.1
        elif user_income <= 1200000:
            tax_rate = 0.15
        elif user_income <= 1500000:
            tax_rate = 0.20
        else:
            tax_rate = 0.30

        # Ensure non-negative taxable income
        taxable_income = max(user_income - deductions - standard_deduction, 0)
        return taxable_income * tax_rate

    # Extract standard deduction from regulations
    standard_deduction = regulations_df['Standard_Deductions'].iloc[0]

    # Apply tax calculations to each user
    financial_df['Estimated_Tax'] = financial_df.apply(
        lambda row: calculate_tax(
            row['Income'],
            row[['HealthInsurance', 'HomeLoan', 'ELSS', 'NPS', 'PPF', 'HouseRent']].sum(),
            standard_deduction
        ), axis=1
    )

    return financial_df
```

```python
# ✅ Install required libraries (only for Google Colab)
!pip install langchain langchain_community chromadb transformers accelerate bitsandbytes

import pandas as pd
import torch
from torch import cuda, float16
import transformers
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
from langchain.docstore.document import Document
from langchain.vectorstores import Chroma
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain.llms import HuggingFacePipeline
from langchain_community.embeddings import HuggingFaceEmbeddings

# ✅ Step 1: Generate Fake Financial Data
def generate_financial_data(num_users=1000):
    """Generates random financial data for users."""
    import numpy as np

    data = {
        'User_ID': np.arange(1, num_users + 1),
        'Income': np.random.uniform(50000, 2000000, num_users),
        'Expenses': np.random.uniform(5000, 50000, num_users),
        'HealthInsurance': np.random.uniform(1000, 5000, num_users),
        'HomeLoan': np.random.uniform(0, 50000, num_users),
        'ELSS': np.random.uniform(0, 50000, num_users),
        'NPS': np.random.uniform(0, 50000, num_users),
        'PPF': np.random.uniform(0, 50000, num_users),
        'HouseRent': np.random.uniform(5000, 25000, num_users),
        'Previous_Tax_Amount': np.random.uniform(10000, 100000, num_users),
        'State': np.random.choice(['KA', 'MH', 'DL', 'TN', 'UP', 'WB'], num_users),
        'Filing_Status': np.random.choice(['Single', 'Married', 'Head of Household'], num_users),
        'Tax_Credits': np.random.uniform(2000, 10000, num_users)
    }
    return pd.DataFrame(data)

# ✅ Step 2: Generate Tax Regulations
def generate_tax_regulations():
    """Creates tax brackets and deductions based on Indian tax laws."""
    tax_data = {
        'Standard_Deductions': [75000],  # Assuming flat deduction for all users
    }
    return pd.DataFrame(tax_data)

# ✅ Step 3: Apply Tax Regulations
def apply_tax_regulations(financial_df, regulations_df):
    """Calculates estimated tax based on income and deductions."""

    def calculate_tax(user_income, deductions, standard_deductions):
        """Simplified tax calculation logic."""
        if user_income <= 300000:
            tax_rate = 0.0
        elif user_income <= 700000:
            tax_rate = 0.05
        elif user_income <= 1000000:
            tax_rate = 0.1
        elif user_income <= 1200000:
            tax_rate = 0.15
        elif user_income <= 1500000:
            tax_rate = 0.20
        else:
            tax_rate = 0.30

        taxable_income = max(user_income - deductions - standard_deductions, 0)
        return taxable_income * tax_rate

    standard_deductions = regulations_df['Standard_Deductions'].iloc[0]
    financial_df['Estimated_Tax'] = financial_df.apply(
        lambda row: calculate_tax(row['Income'], row[['HealthInsurance', 'HomeLoan', 'ELSS', 'NPS', 'PPF', 'HouseRent']].sum(), standard_deductions),
        axis=1
    )
    return financial_df

# ✅ Step 4: Generate Financial Data & Apply Tax Regulations
num_users = 1000
financial_data = generate_financial_data(num_users)
tax_regulations = generate_tax_regulations()
financial_data_with_taxes = apply_tax_regulations(financial_data, tax_regulations)

# ✅ Step 5: Convert Financial Data to LangChain Documents
documents = [
    Document(
        page_content=(
            f"User_ID: {row['User_ID']}, Income: {row['Income']}, Expenses: {row['Expenses']}, "
            f"HealthInsurance: {row['HealthInsurance']}, HomeLoan: {row['HomeLoan']}, "
            f"ELSS: {row['ELSS']}, NPS: {row['NPS']}, PPF: {row['PPF']}, HouseRent: {row['HouseRent']}, "
            f"Previous_Tax_Amount: {row['Previous_Tax_Amount']}, State: {row['State']}, "
            f"Filing_Status: {row['Filing_Status']}, Tax_Credits: {row['Tax_Credits']}, "
            f"Estimated_Tax: {row['Estimated_Tax']}"
        )
    )
    for _, row in financial_data_with_taxes.iterrows()
]

# ✅ Step 6: Initialize ChromaDB
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
langchain_chroma = Chroma.from_documents(documents, embedding_model)

# ✅ Step 7: Load Zephyr-7B Model for LLM Processing
device = f'cuda:{cuda.current_device()}' if cuda.is_available() else 'cpu'
model_id = 'HuggingFaceH4/zephyr-7b-beta'

bnb_config = transformers.BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type='nf4',
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.float16
)

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    device_map='auto',
)
tokenizer = AutoTokenizer.from_pretrained(model_id)

query_pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.float16,
    max_new_tokens=500,
    device_map="auto",
)

llm = HuggingFacePipeline(pipeline=query_pipeline)

# ✅ Step 8: Define Prompt Template & QA Chain
template = """
Based on the following financial data and tax regulations, analyze and provide personalized tax-saving recommendations:
Financial Data: {question}
Context: {context}
Answer:
"""
PROMPT = PromptTemplate(input_variables=["context", "question"], template=template)

retriever = langchain_chroma.as_retriever(search_kwargs={"k": 5})

# ✅ Function to Remove Duplicates
def remove_duplicates(documents):
    seen = set()
    unique_docs = []
    for doc in documents:
        if doc.page_content not in seen:
            unique_docs.append(doc)
            seen.add(doc.page_content)
    return unique_docs

# ✅ Set up RetrievalQA
qa_chain = RetrievalQA.from_chain_type(
    llm, retriever=retriever, chain_type_kwargs={"prompt": PROMPT}
)

def get_tax_optimization_recommendations(query):
    """Retrieves relevant documents and generates tax-saving advice."""
    raw_docs = retriever.get_relevant_documents(query)
    unique_docs = remove_duplicates(raw_docs)
    context = " ".join([doc.page_content for doc in unique_docs])
    result = qa_chain.run(query)
    return result

# ✅ Step 9: Test with an Example Query
query = "Analyze - User_ID: 317, Income: 65185.29, Expenses: 6770.46, HealthInsurance: 1921.03, HomeLoan: 0.0, ELSS: 0.0, NPS: 1767.37, PPF: 1927.76, HouseRent: 3657.13, Previous_Tax_Amount: 15957.37, State: VI, Filing_Status: Head of Household, Tax_Credits: 2990.91, Estimated_Tax: 9660.64"

response = get_tax_optimization_recommendations(query)
print(response)
```
